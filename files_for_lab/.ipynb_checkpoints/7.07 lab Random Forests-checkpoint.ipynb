```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split

numerical = pd.read_csv('numerical.csv')
categorical = pd.read_csv('categorical.csv')
targets = pd.read_csv('target.csv')
data = pd.concat([numerical, categorical, targets], axis = 1)
data['TARGET_B'].value_counts()
```

```python
data.head()
```

```python
data.isna().sum().sum()
```

```python
y = data['TARGET_B']
X = data.drop(['TARGET_B'], axis = 1)
```

#### Split the data into a training set and a test set.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
X_train_num = X_train.select_dtypes(np.number)
X_test_num = X_test.select_dtypes(np.number)
X_train_cat = X_train.select_dtypes(object)
X_test_cat = X_test.select_dtypes(object)
```

#### Scale the features either by using MinMax Scaler

```python
from sklearn.preprocessing import MinMaxScaler
transformer = MinMaxScaler().fit(X_train_num)
# Scaling
X_train_scaled = pd.DataFrame(transformer.transform(X_train_num), columns=X_train_num.columns)
X_test_scaled = pd.DataFrame(transformer.transform(X_test_num), columns=X_test_num.columns)
```

#### Encode the categorical features using One-Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
# Fitting the encoder
encoder = OneHotEncoder(drop='first', handle_unknown = 'ignore').fit(X_train_cat)
column_name = encoder.get_feature_names_out(X_train_cat.columns)
# Encoding
X_train_hot = pd.DataFrame(encoder.transform(X_train_cat).toarray(), columns = column_name)
X_test_hot = pd.DataFrame(encoder.transform(X_test_cat).toarray(), columns = column_name)
```

```python
X_train = pd.concat([X_train_scaled,X_train_hot], axis = 1)
X_test = pd.concat([X_test_scaled,X_test_hot], axis = 1)
```

```python
y_train = y_train.reset_index(drop=True) 
y_test = y_test.reset_index(drop=True)
```

```python
y_train1=y_train
```

```python
X_train_scaled1 = X_train_scaled.drop(['TARGET_D'], axis=1)
```

## Upsampling (oversampling)

```python
trainset = pd.concat([X_train, y_train], axis=1)
```

```python
y_train.value_counts()
```

```python
from sklearn.utils import resample

category_0 = trainset[trainset['TARGET_B'] == 0]
category_1 = trainset[trainset['TARGET_B'] == 1]
```

```python
category_1_oversampled = resample(category_1, 
                                  replace=True, 
                                  n_samples = len(category_0))

print(category_0.shape)
print(category_1_oversampled.shape)
```

```python
trainset_new = pd.concat([category_0, category_1_oversampled], axis = 0)
trainset_new = trainset_new.sample(frac =1)

X_train = trainset_new.drop(['TARGET_B'], axis=1)
y_train = trainset_new['TARGET_B']
#data = data.reset_index(drop=True)
print(X_train.shape)
```

```python
X_train = pd.DataFrame(X_train)
X_test = pd.DataFrame(X_test)

y_train_regression = X_train['TARGET_D']
y_test_regression = X_test['TARGET_D']

# Now we can remove the column target d from the set of features
X_train = X_train.drop(['TARGET_D'], axis = 1)
X_test = X_test.drop(['TARGET_D'], axis = 1)
```

## Random Forests algorithm

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

clf = RandomForestClassifier(max_depth=5,
                             min_samples_split=20,
                             min_samples_leaf =20,
                             max_samples=0.8,
                             random_state = 42)
clf.fit(X_train, y_train)
print(clf.score(X_train, y_train))
print(clf.score(X_test, y_test))

y_pred = clf.predict(X_test)
display(y_test.value_counts())
display(confusion_matrix(y_test, y_pred))

# max_depth=5,   
# min_samples_split=20,
# min_samples_leaf =20,    ultimate anser based on at least this many rows
# max_samples=0.8,   fraction of X-train to use in each tree
# random_state = 42)   
```

## Feature Selections 

#### Variance Threshold

```python
from sklearn.feature_selection import VarianceThreshold  
var_threshold = 0.02
sel = VarianceThreshold(threshold=(var_threshold))
```

```python
y = data['TARGET_B']
X = data.drop(['TARGET_B','TARGET_D'], axis = 1)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

X_train_num = X_train.select_dtypes(np.number)
X_test_num = X_test.select_dtypes(np.number)
X_train_cat = X_train.select_dtypes(object)
X_test_cat = X_test.select_dtypes(object)
```

```python
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train_num)
X_test_scaled = scaler.fit_transform(X_test_num)
```

```python
X_train_scaled.shape
```

```python
X_test_scaled.shape
```

```python
# This drops the columns that have a variance less than this threshold
sel = sel.fit(X_train_scaled)
temp_train = sel.transform(X_train_scaled)
temp_train = pd.DataFrame(temp_train)
print(X_train_scaled.shape)
print(temp_train.shape)
```

```python
pd.DataFrame(temp_train)
```

```python
sel.variances_ > var_threshold
sel.get_support()
var_list = list(sel.get_support())
col_list=[col[0] for col in zip(X_train_num.columns, var_list) if col[1] == True]
```

```python
len(col_list)
```

```python
X_test_num = X_test_num.loc[:, col_list]
X_train_cat = X_train_cat.reset_index(drop=True) 
```

```python
# Fitting the encoder
encoder = OneHotEncoder(drop='first', handle_unknown = 'ignore').fit(X_train_cat)
column_name = encoder.get_feature_names_out(X_train_cat.columns)
# Encoding
X_train_hot = pd.DataFrame(encoder.transform(X_train_cat).toarray(), columns = column_name)
X_test_hot = pd.DataFrame(encoder.transform(X_test_cat).toarray(), columns = column_name)

X_train_hot = X_train_hot.reset_index(drop=True) 
temp_train = temp_train.reset_index(drop=True)
X_test_num = X_test_num.reset_index(drop=True) 

X_train = pd.concat([temp_train,X_train_hot],axis=1)
X_test = pd.concat([X_test_num,X_test_hot],axis=1)

y_train = y_train.reset_index(drop=True) 
y_test = y_test.reset_index(drop=True)
```

```python
# find overlapping column names
overlap_cols = set(temp_train.columns) & set(X_train_hot.columns)

# drop overlapping columns from temp_train
temp_train_drop = temp_train.drop(columns=overlap_cols)

# concatenate horizontally
X_train = pd.concat([temp_train_drop, X_train_hot], axis=1)
```

```python
trainset = pd.concat([X_train, y_train], axis=1)

category_0 = trainset[trainset['TARGET_B'] == 0]
category_1 = trainset[trainset['TARGET_B'] == 1]

category_1_oversampled = resample(category_1, 
                                  replace=True, 
                                  n_samples = len(category_0))

trainset_new = pd.concat([category_0, category_1_oversampled], axis = 0)
trainset_new = trainset_new.sample(frac =1)

X_train = trainset_new.drop(['TARGET_B'], axis=1)
y_train = trainset_new['TARGET_B']

X_train = pd.DataFrame(X_train)
X_test = pd.DataFrame(X_test)
```

```python
clf = RandomForestClassifier(max_depth=5,
                             min_samples_split=20,
                             min_samples_leaf =20,
                             max_samples=0.8,
                             random_state = 42)
clf.fit(X_train, y_train)
print(clf.score(X_train, y_train))
print(clf.score(X_test, y_test))

y_pred = clf.predict(X_test)
display(y_test.value_counts())
display(confusion_matrix(y_test, y_pred))
```
